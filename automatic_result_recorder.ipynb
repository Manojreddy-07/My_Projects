```python
def openDriver():
    driver = "C:/Users/Lenovo/chromedriver"
    browser = webdriver.Chrome(driver)
    return browser
```

```python
def login(browser,vtu):
    browser.get(('http://exams.veltech.edu.in/Studentlogin/StuLogin.aspx'))
    tempUser = vtu.upper()
    username = browser.find_element(By.ID,'txtUserName')
    username.send_keys(tempUser)
    userPassword = browser.find_element(By.ID, 'txtPassword')
    userPassword.send_keys(tempUser)
    nextButton = browser.find_element(By.ID, 'LoginButton')
    nextButton.click()
    return browser;
```

```python
def invokeResult(browser,semno,tempUser,path):
    result = browser.find_element(by = By.LINK_TEXT, value = 'Results')
    result.click()
    regularresult = browser.find_element(By.XPATH, value = "/html/body/form/div[3]/div/div[2]/ul/li[5]/ul/li[1]/a")
    browser.get((regularresult.get_attribute('href')))
    select = Select(browser.find_element(By.ID,'ContentPlaceHolder1_ddlSemester'))
    sem = browser.find_element(By.ID,'ContentPlaceHolder1_ddlSemester')
    sem.click()
    semno = str(semno)
    select.select_by_value(semno)
    var = tempUser+"sem"+semno+ ".png"
    browser.execute_script("window.scrollTo(0,250)")
    marksheet = browser.get_screenshot_as_file(var)
    print(marksheet)
    #table = browser.find_element(By.ID,'ContentPlaceHolder1_gvExamResult2013')
    #return table;
```

```python
def getResultData(table):

    table_html = table.get_attribute('innerHTML')
    # Parse the HTML table
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(table_html, 'html.parser')

# Find all table rows
    rows = soup.find_all('tr')

# Initialize the result list
    result_list = {}

# Iterate over the rows and extract data
    for row in rows:
        # Find all table cells in the row
        cells = row.find_all('td')

    # Extract the text from each cell and append it to the result list
        cell_data = [cell.get_text(strip=True) for cell in cells]
        if(len(cell_data)<4):
            continue
        else:
            result_list[cell_data[2]] = cell_data[4]
    return result_list
```

```python
def calculateGPA(result_grade):
    score = 0
    credit = 0
    for course in result_grade:
        credit = credit + data[course]
        if(result_grade[course]=='S'):
            score = score + data[course]*10
        elif(result_grade[course]=='A'):
            score = score + data[course]*9
        elif(result_grade[course]=='B'):
            score = score + data[course]*8
        elif(result_grade[course]=='C'):
            score = score + data[course]*7
        elif(result_grade[course]=='D'):
            score = score + data[course]*6
        else:
            continue
    gpa = score/credit;
    return [gpa,credit,score]
```

```python
def calculateCGPA(gpa_data,cgpa_data):
    cgpa_data[2] = cgpa_data[2] + gpa_data[2]
    cgpa_data[1] = cgpa_data[1] + gpa_data[1]
    cgpa_data[0] = cgpa_data[2]/cgpa_data[1]
    return cgpa_data
```
